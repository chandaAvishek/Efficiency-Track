# **INTRODUCTION**
## Objective
The objective of this report is to examine the connection between social media usage, screen time, and individual productivity, with a focus on determining whether increased time spent on these platforms negatively impacts how effectively people work or study. In today’s digital era, social media has become a central part of communication, entertainment, and information exchange, but it can also act as a significant distraction. By understanding these patterns, we aim to identify actionable steps that individuals can take to improve their time management and concentration. This knowledge can also help businesses enhance workplace efficiency and support policymakers in developing guidelines that promote balanced technology use, ultimately benefiting personal well-being and societal productivity.

##Dataset Introduction
This study uses the Social Media vs. Productivity dataset, compiled by Mahdi Mashayekhi and made publicly available on Kaggle. The dataset contains survey-based information capturing individuals’ daily habits, including hours spent on social media, total screen time, frequency of notifications, and self-reported productivity levels. Provided in a structured, user-friendly CSV format, it is both compact and comprehensive, making it easy to work with for analysis. Its relevance lies in the direct alignment between its recorded variables and the study’s objective, as it allows for the exploration of measurable correlations between digital behavior and productivity outcomes. This real-world data makes it possible to produce insights that are grounded in actual user experiences.

##Problem Statement
As social media becomes increasingly integrated into daily life, the question of whether its growing presence undermines productivity has become a pressing concern. Many individuals struggle to balance the benefits of connectivity with the risks of distraction, and organizations face challenges in maintaining focus and efficiency among employees. However, quantifying the exact impact of screen time on productivity is complex, as it involves subjective perceptions, varying work habits, and diverse personal circumstances. This report addresses the problem by analyzing empirical data to identify patterns and potential negative effects of excessive social media use, aiming to separate anecdotal assumptions from evidence-based conclusions.

# **Feature** **Extraction**

**Social Media Intensity Index**: *daily_social_media_time gives hours, ignoring notifications of variety of platforms. By z-score scaling and summing components we get a single intensity indicator*

````python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
social_media_features = ['daily_social_media_time', 'number_of_notifications']
social_data['social_media_intensity'] = scaler.fit_transform(social_data[social_media_features]).sum(axis = 1)
````

This feature captures "how long" and "how interrupted" a user is by social media



Social Media Time vs. Proportion of Work Hours: Spending 3 hours online means different things for someone working 12 hours (low proportion) vs 5 hours (high proportion)

````python
social_data['sm_time_work_ratio'] = social_data['daily_social_media_time'] / (social_data['work_hours_per_day'] + 1e-6)
````

This feature measures how much social media use intrudes on available work time

# Platform Category Encoding: Different platform have different distractions.

````python
platform_map = {
    'Facebook': 'entertainment',
    'Instagram': 'entertainment',
    'TikTok': 'entertainment',
    'Telegram': 'messaging',
    'Twitter': 'news'
}

social_data['platform_category'] = social_data['social_platform_preference'].map(platform_map)
platform_dummies = pd.get_dummies(social_data['platform_category'], prefix = 'plat')
social_data = pd.concat([social_data, platform_dummies], axis=1)
````

This feature allows the model to predict if certain types of platforms disrupts productivity

# Hypothesis Testing:

# Separate Hypothesis Testing and Predictive Modeling

 Hypothesis statement: Does higher daily social media time lead to lower actual productivity score?

 Null Hypothesis (H0): There is no significant difference in actual productivity score between users with high and low daily social media time.

 Alternative Hypothesis (H1): There is significant difference in actual productivity score between users with high and low daily social media time.

# Choice of test statistics: Independent samples t-test

*  The t-test is used to compare the means of two independent groups.
* Here high vs. low users daily use of social media time, splits at the median

# The t-test statistics is appropiate since:
* The response variable actual_productivity_score is continuous
* The independent group is categorical
* The goal is to access whether the high and the low groups have significantly different means.

````python
# Spillting the data based on the median

median_time = social_data['daily_social_media_time'].median()
group_high = social_data[social_data['daily_social_media_time'] > median_time]['actual_productivity_score']
group_low = social_data[social_data['daily_social_media_time'] <= median_time]['actual_productivity_score']
````

# Decision: If the p-value, p_val <= 0.05 we reject the null hypothesis (H0) otherwise we fail reject the null hypothesis (H0)

````python
# applying t-test statistics

import scipy.stats as stats

t_stats, p_val = stats.ttest_ind(group_high, group_low, equal_var=False)
print(f'T-statistics: {t_stats:.3f}, P-value: {p_val:.3f}')
````

 From the hypothesis test we get the t-statistics and p-value:
**T-statistics: -1.089, P-value: 0.276**

 Conclusion: Since the p_val = 0.276 > 0.05 we fail to reject the null hypothesis (H0). This means there is not enough statistical evidence to say there is significant difference in actual productivity scores between high and low socail media users, based on our data.

 Hypothesis statement: Is there a significant difference in actual productivity between users who have digital wellbeing tools enabled and those who do not?

 Null Hypothesis (H0): There is no significant difference in actual productivity scores between users with digital wellbeing tools enabled.

 Alternative Hypothesis (H1): There is a significnt difference in actual productivity scores between users with digital wellbeing tools enabled.

# Choice of test statistics: Independent samples t-test.
* The test is used to compare the means of two independent groups, on a continuous outcome.

````python
# splitting the data
group_enabled = social_data[social_data['has_digital_wellbeing_enabled'] == True]['actual_productivity_score']
group_disabled = social_data[social_data['has_digital_wellbeing_enabled'] == False]['actual_productivity_score']
````

# Decision: If the p-value, p_val <= 0.05 we reject the null hypothesis (H0) otherwise we fail reject the null hypothesis (H0)

````python
# test statistics

from scipy.stats import ttest_ind
t_stats , p_val = ttest_ind(group_enabled, group_disabled, equal_var = False)
print(f'T-satistics: {t_stats:.3f}, P-value: {p_val:.3f}')
````

 From the hypothesis test we get: **T-satistics: -1.145, P-value: 0.252**

 Conclusion: Since the p_val = 0.252 > 0.05 we fail to reject the null hypothesis (H0). This means there is not enough statistical evidence to say there is significant difference in actual productivity scores between users with digital wellbeing tools enabled, based on our data.

# Regression Analysis

 1. Data Splitting Strategy: We split our dataset to ensure that our model performance is evaluated on unseen data, helping to detect overfitting and estimate generalization performance. We use a random 80/20 splits (80% training, 20% testing) via sklearn's ```train_test_split```.

````python
from sklearn.model_selection import train_test_split
social_data_encoded = pd.get_dummies(social_data, drop_first=True)
x = social_data_encoded.drop(['actual_productivity_score'], axis = 1)
y = social_data_encoded['actual_productivity_score']

# test size = 20%
# random_state = 42 ensures reproducibility
# Splitting ensures unbiased model on unseen data
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
````

# Mathematical Expression

# 1. Linear Regression Model: We have used linear regression model, because it shows linear relationship between inputs and productivity score.


```
y_hat = β₀ + β₁X₁ + β₂X₂ + ... + βₙXₙ
```





````python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, Y_train)

predictions = model.predict(X_test)
````

````python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(Y_test, predictions)
r2 = r2_score(Y_test, predictions)
print(f'Mean Squared Error: {mse:.3f}')
print(f'R-squared: {r2:.3f}')
````

The linear regression model explains approximately 83% of the variability in actual productivity based on behavioral variables (screen time, notifications, etc.).
The mean square error (MSE) is relatively low (0.54), which means that the predictions are fairly close to the actual values.

# The importance of variables

````python
coefs = pd.Series(model.coef_, index=x.columns)

coefs_sorted = coefs.sort_values(ascending=False)

coefs_sorted
````

The factor that contributes most to predicting actual productivity is the productivity perceived by the user themselves.
This is a good subjective indicator, recognized by the model.
The second most important factor is the level of job satisfaction.
Users who are more satisfied with their jobs tend to be more productive, which is consistent with the literature on well-being at work.
The gender_Other variable, on the other hand, has a very weak and negative effect on predicted productivity.

### However, linear regression assumes a linear relationship between the variables and the target, which may not reflect reality in more complex data. This is why a second model was considered.

# 2. Random Forest : The Random Forest model is capable of modeling complex, non-linear relationships.

````python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(
    n_estimators=100,
    max_depth=None,
    random_state=42
)
rf_model.fit(X_train, Y_train)
````

## Explanation of how hyperparameters were chosen

Unlike linear regression, the Random Forest model has several hyperparameters (IP parameters) that influence its performance:

*   n_estimators: number of trees to build
*   max_depth: maximum depth of each tree
*   Random_state : is used to check the reproducibility of results when random operations are involved.



````python
#Predictions
y_train_pred_rf = rf_model.predict(X_train)

y_test_pred_rf = rf_model.predict(X_test)
````

````python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

#Model Evaluation

rmse_train = np.sqrt(mean_squared_error(Y_train, y_train_pred_rf))
mae_train = mean_absolute_error(Y_train, y_train_pred_rf)
r2_train = r2_score(Y_train, y_train_pred_rf)


rmse_test = np.sqrt(mean_squared_error(Y_test, y_test_pred_rf))
mae_test = mean_absolute_error(Y_test, y_test_pred_rf)
r2_test = r2_score(Y_test, y_test_pred_rf)

print(" Performance on the training set :")
print(f"RMSE : {rmse_train:.3f}, MAE : {mae_train:.3f}, R² : {r2_train:.3f}")

print("\n Performance on the test set :")
print(f"RMSE : {rmse_test:.3f}, MAE : {mae_test:.3f}, R² : {r2_test:.3f}")
````

The Random Forest model used with the hyperparameters (n_estimators=100, max_depth=None, random_state=42) produced reliable, stable, and accurate predictions.


*   The R² values close to 1 for both sets (0.998 and 0.989) indicate that the model explains almost all of the variance in the target variable (productivity).
*   The RMSE and MAE errors are low, which means that the difference between the predictions and the actual values is minimal.



These results show that the model is well suited to the problem of predicting productivity.

#**CONCLUSION**

The analysis of the Social Media vs. Productivity dataset reveals notable relationships between the intensity of social media use, prolonged screen time, and decreased self-reported productivity. Findings suggest that while these platforms may facilitate information exchange and professional networking, they also represent a substantial source of distraction, potentially undermining concentration and efficiency when their use is not adequately regulated.

The implications are twofold. On an individual level, the results highlight the need for strategies to manage digital engagement, such as disabling non-essential notifications, scheduling specific time slots for social media interaction, and implementing periods of intentional disconnection. On an organizational level, the findings underscore the importance of internal policies that limit digital interruptions and promote work environments conducive to sustained focus.

It is important to acknowledge the limitations of this study, particularly the self-reported nature of the data, which may introduce perceptual biases, and the absence of objective productivity metrics. Future research could benefit from longitudinal data, experimental designs, and more diverse samples to enhance the robustness and generalizability of the conclusions.

Ultimately, this study supports the hypothesis that unregulated social media usage can hinder productivity while offering practical avenues for harmonizing digital connectivity with performance. Conscious and deliberate management of these tools emerges as a key competency in the modern era, at the intersection of technological, economic, and societal challenges.
